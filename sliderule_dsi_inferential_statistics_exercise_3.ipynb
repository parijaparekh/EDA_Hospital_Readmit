# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

##### Null Hypothesis
Hospital discharge numbers and readmission rates are not correlated.  

##### Alternate Hypothesis
Hospital discharge numbers and readmission rates are correlated

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def pearson_r(x, y):
    corr_mat = np.corrcoef(x,y)
    return corr_mat[0,1]
```

```python
df = pd.read_csv("data/cms_hospital_readmissions.csv")
analysis_info = df.loc[:][['Number of Discharges', 'Excess Readmission Ratio']]

## cleeaning up the data:
analysis_info.dropna(inplace = True, axis = 0)
analysis_info = analysis_info[analysis_info['Number of Discharges'] != 'Not Available']
analysis_info['Number of Discharges'] = analysis_info['Number of Discharges'].astype('int')
#print(analysis_info.info())
## Since discharges is non-null object wherein we expected int or float, we check for its values.

hospital_discharges = analysis_info['Number of Discharges']
rate_of_readmissions = analysis_info['Excess Readmission Ratio']

## getting the bootstrap permutted samples for hospital_discharge. 
pearson_for_samples = np.empty(10000)

for i in range(10000):
    pearson_for_samples[i] = pearson_r(np.random.permutation(hospital_discharges), rate_of_readmissions)

print(pearson_r(hospital_discharges, rate_of_readmissions))
## Since the correlation is negative we check the probability of atleast as low as the samples pearson correlation
p_value = np.sum(pearson_for_samples < pearson_r(hospital_discharges, rate_of_readmissions))/len(pearson_for_samples)   

print(p_value)
```

#####  Since the p_value is 0.0 which is less than 0.01 (statistical significance 𝛼 = .01), we reject the Null Hypothesis. 
##### We accept the alternate hypothesis: 
 There is correlation between number of discharges and excess rate of readmissions at a hospital.

##### Yes i do agree with the above analysis and recommendations. 
As it is very evident from the scatter plot that there are more cases of recurrent readmissions happening in institutes that have low number of discharges. 
The pearson correlation between number of discharges and excess rate of readmissions -0.097. This value indicates that both these values are negatively correlated. A lower value for pearson correlation does indicate that the relationship is not linear. This is even evident from the scatter plot. 
To understand how the strength of the correlation coefficient changes with increase in number of beds, we slice the dataframe based on the number of beds staring from 10 to 500.  

```python
correlation_df = pd.DataFrame()

for i in range(10,500,25):
    df_less_than_500 = analysis_info[analysis_info['Number of Discharges'] <= i]
    d = {'number of discharges': [i], 'count' : [df_less_than_500.count()], 
         'correlation': [pearson_r(df_less_than_500['Number of Discharges'], df_less_than_500['Excess Readmission Ratio'])]}
    correlation_df = correlation_df.append(pd.DataFrame(data=d), ignore_index=True)

sns.set()
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(correlation_df['number of discharges'], correlation_df['correlation'], marker=".")
ax.set_xlabel('Number of Discharges')
ax.set_ylabel('Correlation')
ax.set_title('Correlation w.r.t Number of Discharges')
plt.show()
```

##### As we can observe in the graph above, the strength of the correlation gradually drops as the Number of Discharges increases
